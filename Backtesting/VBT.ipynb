##Sinais de Trading

###Cross Above

```python
#Capturar os dados

IBOV_ohlc = vbt.YFData.download('^BVSP', start='2022-01-01', end='2022-06-30', interval='1d').get(['Open', 'High','Low', 'Close', 'Volume'])
IBOV_ohlc
```

```python
#Adicionar o indicador

IBOV_ohlc['MMA_5p'] = vbt.MA.run(IBOV_ohlc.Close, 5).ma
```

```python
#Remover os nans

IBOV_ohlc.dropna(inplace=True)
```

```python
IBOV_ohlc
```

```python
#Quando o preço cruzar acima da média móvel ele me dará um indicador de True

IBOV_ohlc['Cross Above'] = IBOV_ohlc['Close'].vbt.crossed_above(IBOV_ohlc['MMA_5p'])
IBOV_ohlc
```

###Cross Below

```python
IBOV_ohlc['Cross Below'] = IBOV_ohlc['Close'].vbt.crossed_below(IBOV_ohlc['MMA_5p'])
IBOV_ohlc
```

###And e Or

```python
dolar = vbt.YFData.download('USDBRL=X', start='2021-01-01', end ='2022-01-01', interval='1d').get(['Open', 'High', 'Low', 'Close', 'Volume'])
MMAs_dolar = vbt.MA.run(dolar.Close,[9,20,100])
IFR_dor = vbt.RSI.run(dolar.Close, 14)
```

```python
#Estratégia quando a MM9 estiver acima da MM20 E a MM20 estiver acima da MM100

MMAs_dolar[9].ma_above(MMAs_dolar[20]).vbt.signals.AND(MMAs_dolar[20].ma_above(MMAs_dolar[100]))
```

```python
#Quantos sinais eu tive?

MMAs_dolar[9].ma_above(MMAs_dolar[20]).vbt.signals.AND(MMAs_dolar[20].ma_above(MMAs_dolar[100])).sum()
```

```python
#Visualizar a estratégia

fig_MM_all = dolar.vbt.ohlcv.plot()
MMAs_dolar.ma.vbt.plot(fig=fig_MM_all).show()
```

```python
#Gerar o sinal quando a MM9P estiver acima da MM20P ou MM20P estiver acima da MM100P

MMAs_dolar[9].ma_above(MMAs_dolar[20]).vbt.signals.OR(MMAs_dolar[20].ma_above(MMAs_dolar[100])).sum()
```

###Quantificar sequência de sinais

```python
#Quero saber quantas vezes o preço do bitcoin fechou acima da média móvel de 9 períodos no mês de março

btc_mar = vbt.YFData.download('BTC-USD', start='2022-03-01 BRT', end='2022-03-31 BRT', interval='1d').get(['Close'])
btc_mar['EMA9P'] = vbt.talib('EMA').run(btc_mar.Close,9).real
btc_mar.dropna(inplace=True)
```

```python
btc_mar
```

```python
btc_mar['Close_acima_MM'] = btc_mar.Close >=btc_mar.EMA9P
btc_mar
```

```python
#Sequencia acumulada

btc_mar.Close_acima_MM.vbt.signals.pos_rank()
```

```python
btc_mar.Close_acima_MM.vbt.signals.partition_ranges().duration.values
```

##Backtesting

###Cruzamento de preço e média móvel

```python
#Definindo o ativo

mglu = vbt.YFData.download('MGLU3.SA', start='2012-01-01', end='2022-06-30').get('Close')
mglu
```

```python
#Calcular a média móvel

MMA50p = vbt.MA.run(mglu, 50, short_name='MMA').ma
```

```python
#Definir regras de entradas e saídas

entradas = mglu.vbt.crossed_above(MMA50p)
saidas = mglu.vbt.crossed_below(MMA50p)
entradas
```

```python
#Portfolio

portfolio_mglu = vbt.Portfolio.from_signals(mglu,
                                            entradas,
                                            saidas,
                                            fees=0.005,
                                            size=np.inf,
                                            direction='both',
                                            freq='D')
```

```python
#Métricas
portfolio_mglu.stats()
```

```python
#Gráficos

portfolio_mglu.plot(width=500, height=700).show()
```

```python
portfolio_mglu.plot(subplots=['cash', 'assets', 'value'], width=500, height=700).show()
```

```python
fig = mglu.vbt.plot(trace_kwargs=dict(name='Close'))
MMA50p.vbt.plot(trace_kwargs=dict(name='MMA'), fig=fig)
portfolio_mglu.positions.plot(fig=fig).show()
```

```python
#Explorar as ordens

portfolio_mglu.orders.records_readable
```

###Portfólio

```python
lista_ativos = ['BRML3.SA', 'PETR4.SA', 'WEGE3.SA', 'VALE3.SA', 'NTCO3.SA']
```

```python
carteira = vbt.YFData.download(lista_ativos, start= '2019-01-01', end='2021-12-31', missing_index='drop').get('Close')
```

```python
carteira
```

```python
#Criar uma lista com os pesos

alocacao = pd.DataFrame.vbt.empty_like(carteira, fill_value=np.nan)
alocacao.iloc[0] =[0.30, 0.10, 0.30, 0.10, 0.20]
```

```python
#Construir o portfolio
#https://vectorbt.dev/api/portfolio/enums/#vectorbt.portfolio.enums.SizeType

portfolio = vbt.Portfolio.from_orders(
    close = carteira,
    size = alocacao,
    size_type= 'targetpercent',
    group_by=True,
    cash_sharing=True,
    init_cash=1000,
    freq = 'd'

)
```

```python
portfolio.stats(settings=dict(risk_free = 0, freq='d'))
```

```python
portfolio.returns_stats(settings=dict(risk_free = 0, freq='d'))
```

```python
portfolio.plot().show()
```

```python
portfolio.drawdowns.plot(top_n=3).show()
```

```python
portfolio.plot_underwater().show()
```

##Otimização de Parâmetros

```python
ativo_close = vbt.YFData.download('MGLU3.SA', start='2012-01-01', end= '2022-06-30', missing_index='drop').get('Close')
```

```python
#Criar uma lista de médias móveis

MMAs_5_200p = list(range(5, 201))
str(MMAs_5_200p)
```

```python
MMAs_5_200p_df = vbt.MA.run(ativo_close, MMAs_5_200p, short_name='MMAs').ma
```

```python
type(MMAs_5_200p_df)
```

```python
#Definir entradas e saídas

entrada = ativo_close.vbt.crossed_above(MMAs_5_200p_df)
saidas = ativo_close.vbt.crossed_below(MMAs_5_200p_df)
```

```python
#Construir o portfólio

pf_cruz_pre_MMA = vbt.Portfolio.from_signals(ativo_close,
                                             entradas,
                                             saidas,
                                             fees=0.005,
                                             direction='both',
                                             freq='d')
```

```python
#Gráfico de Retornos

pf_cruz_pre_MMA.total_return().vbt.plot().update_yaxes(title_text='<b> Retorno total').update_xaxes(title_text='<b> MM do Cruzamento').show()
```

```python
pf_cruz_pre_MMA.trades.profit_factor().vbt.plot().show()
```

```python
pf_cruz_pre_MMA.drawdowns.max_drawdown().vbt.plot().show()
```
